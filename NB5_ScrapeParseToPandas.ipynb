## Scrape the Billboard Hot 100 with requests

https://www.billboard.com/charts/hot-100


## Parse with Beautifulsoup and 
## Push the Data to a Pandas Dataframe

*   Import the necessary packages
*   Scrape the site
*   Get the html
*   Turn the html into soup






```
# We will use pandas and datetime as we progress towards building the dataframe
import pandas as pd
import datetime
today = str(datetime.datetime.now().date())

# We will need requests and BeautifulSoup for scraping and parsing (respectively).
import requests 
from bs4 import BeautifulSoup

# Scrape (get) the website for which you have interest
response = requests.get('https://www.billboard.com/charts/hot-100')

# Turn the response into a readable object
html = response.content

# Turn the html into Soup ... an object that can be more easily accessed, read and parsed!
soup = BeautifulSoup(html, 'lxml')
```

### Loop through all of the 'songs' ... and create a complete outline of the data you plan to collect assigning each data element a temporary string until it can be replaced by the actual data. Example: title = 'This is the position title'. End the loop by appending the new row (as a dictionary) to 'chart'. Finally, once the loop is completed create a dataframe 'Hot100' from the list 'chart'.
 

```
chart = []

for songs in soup.find_all('li',{'class':'chart-list__element'}):
  song = songs.find('span',{'class':'chart-element__information__song'}).string
  artist = 'This is the artist'
  rank = 'This is the rank'
  trending = 'Trending'
  change = 'Change from last week'
  lastWeek = 'Rank from last week'
  peak = 'Best ranking'
  duration = 'Weeks on the chart'
  imageLink = 'This is the link to the image'
  date = today

  chart.append({'Song': song, 'Artist': artist, 'Rank': rank, 'Trending': trending, 
                'Change': change, 'LastWeek': lastWeek, 'Peak': peak, 'Duration': duration,
                'ImageLink': imageLink, 'Date': date})

Hot100 = pd.DataFrame(chart)
Hot100
```
